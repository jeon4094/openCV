```python
from google.colab import drive
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/drive/MyDrive/Colab Notebooks/Images/girl.jpg')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
G_img = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(G_img.shape)
plt.imshow(G_img, cmap='gray')
```

```python
gPlus = np.zeros((G_img.shape[0], G_img.shape[1]), dtype = np.ubyte)
val = 50
for h in range(G_img.shape[0]):
  for w in range(G_img.shape[1]):
    plusVal = G_img[h,w] + val

    if(plusVal > 255):
      plusVal = 255
    gPlus[h,w] = plusVal
plt.imshow(gPlus, cmap = 'gray')
```

```python
img = cv2.imread('/content/drive/MyDrive/Colab Notebooks/Images/aero2.bmp')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gimg = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gimg.shape)
#plt.imshow(gimg, cmap='gray')
cv2_imshow(gimg)
```

```python
img = cv2.imread('/content/drive/MyDrive/Colab Notebooks/Images/lena256.jpg')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gimg2 = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gimg2.shape)
#plt.imshow(gimg2, cmap='gray')
cv2_imshow(gimg2)
```

```python
for h in range(G_img.shape[0]):
  for w in range(G_img.shape[1]):
    plusVal = (np.int32(gimg[h,w] + gimg2[h,w]))/2

    if(plusVal > 255):
      plusVal = 255
    gPlus[h,w] = plusVal
#plt.imshow(gPlus, cmap = 'gray')
cv2_imshow(gPlus)
```

```python
weight = 0.3

for h in range(G_img.shape[0]):
  for w in range(G_img.shape[1]):
    plusVal = np.int32(gimg[h,w]*weight + gimg2[h,w])*(1-weight)

    if(plusVal > 255):
      plusVal = 255
    gPlus[h,w] = plusVal
#plt.imshow(gPlus, cmap = 'gray')
cv2_imshow(gPlus)
```
